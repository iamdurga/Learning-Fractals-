# The Mandelbrot Set

## Introduction

The Mandelbrot set is a well-known example of a fractal in complex dynamics. It is defined by a particular iterative process involving complex numbers, and it exhibits intricate and self-similar structures.

## Mathematical Definition

The Mandelbrot set is defined by the iterative function:

$$ z_{n+1} = z_n^2 + c $$

where $ z $ and $ c $ are complex numbers and the sequence starts with $ z_0 = 0 $.

A complex number $ c $ is part of the Mandelbrot set if the sequence $z_n$ remains bounded as $n $ approaches infinity. Formally:

$$ M = \left\{ c \in \mathbb{C} : \forall n \in \mathbb{N}, |z_n| \leq 2 \right\} $$

## Iterative Process

1. **Initialization**: Start with $ z_0 = 0 $.
2. **Iteration**: Compute $ z_{n+1} $ using the formula $z_{n+1} = z_n^2 + c $.
3. **Escape Condition**: Check if the magnitude of $ z_n $ exceeds 2. If it does, the sequence is unbounded, and $ c $ is not part of the Mandelbrot set.
4. **Repeat**: Continue iterating until the maximum number of iterations is reached or the escape condition is met.



## Example: $ c = 0.3 + 0.5i $

Let's determine if the complex number $ c = 0.3 + 0.5i $ is part of the Mandelbrot set by following the iterative process.

1. **Initialization**: Start with $ z_0 = 0 $.

2. **Iteration**: Compute $ z_{n+1} = z_n^2 + c $ for each $ n $.

Let's go through the first few iterations:

### Iteration 0:
$ z_0 = 0 $

### Iteration 1:
$ z_1 = z_0^2 + c = 0^2 + (0.3 + 0.5i) = 0.3 + 0.5i $

### Iteration 2:
$$
 \begin{aligned}
& z_2 = z_1^2 + c = (0.3 + 0.5i)^2 + (0.3 + 0.5i) \\
& z_2 = (0.09 + 0.3i + 0.25i^2) + (0.3 + 0.5i) \\
& z_2 = (0.09 + 0.3i - 0.25) + (0.3 + 0.5i) \\
& z_2 = (-0.16 + 0.3i) + (0.3 + 0.5i) \\
& z_2 = 0.14 + 0.8i \\
\end{aligned}
 $$

### Iteration 3:
$$
\begin{aligned}
z_3 &= z_2^2 + c = (0.14 + 0.8i)^2 + (0.3 + 0.5i) \\
    &= (0.0196 + 0.224i + 0.64i^2) + (0.3 + 0.5i) \\
    &= (0.0196 + 0.224i - 0.64) + (0.3 + 0.5i) \\
    &= (-0.6204 + 0.224i) + (0.3 + 0.5i) \\
    &= -0.3204 + 0.724i
\end{aligned}
 $$

### Iteration 4:
$$
 \begin{aligned}
&z_4 = z_3^2 + c = (-0.3204 + 0.724i)^2 + (0.3 + 0.5i) \\
&z_4 = (0.10265616 - 0.4647296i + 0.524576i^2) + (0.3 + 0.5i) \\
&z_4 = (0.10265616 - 0.4647296i - 0.524576) + (0.3 + 0.5i) \\
&z_4 = (-0.42191984 - 0.4647296i) + (0.3 + 0.5i) \\
&z_4 = -0.12191984 + 0.0352704i
\end{aligned}
 $$

### Iteration 5:
$$
\begin{aligned}
&z_5 = z_4^2 + c = (-0.12191984 + 0.0352704i)^2 + (0.3 + 0.5i) \\
&z_5 = (0.014861056 + 0.0085872848i + 0.001244i^2) + (0.3 + 0.5i) \\
&z_5 = (0.014861056 + 0.0085872848i - 0.001244) + (0.3 + 0.5i) \\
&z_5 = (0.013617056 + 0.0085872848i) + (0.3 + 0.5i) \\
&z_5 = 0.313617056 + 0.5085872848i
\end{aligned}
$$


We continue this iterative process until we either determine that the sequence \( \{z_n\} \) remains bounded or diverges.

### Conclusion for $ c = 0.3 + 0.5i $

As we iterate, we notice that the magnitudes of $ z_n $ do not exceed 2 within these few steps. However, to make a definitive conclusion, we would typically iterate up to a predefined maximum number of iterations (e.g., 1000) and check if the magnitude of $ z_n $ ever exceeds 2.

In practice, for $ c = 0.3 + 0.5i $, the magnitude of $ z_n $ does eventually exceed 2, meaning $ c $ is not part of the Mandelbrot set.




## Escape Condition

The escape condition for the Mandelbrot set is based on the magnitude of $ z_n $. If $ |z_n| > 2 $ for any $ n $, the sequence will diverge to infinity, and the corresponding complex number $ c $ is not in the Mandelbrot set.

## Properties of the Mandelbrot Set

### Self-Similarity

The boundary of the Mandelbrot set is a fractal, displaying self-similar structures at various scales. This means that zooming into the boundary reveals smaller copies of the Mandelbrot set, often distorted but recognizably similar to the whole set.

### Fractal Dimension

The Mandelbrot set has a fractal dimension, which is a measure of its complexity. The boundary of the Mandelbrot set is infinitely intricate and cannot be described by an integer dimension. Instead, it has a fractional dimension, indicating its fractal nature.

### Connectedness

The Mandelbrot set is connected, meaning there is a path between any two points in the set that lies entirely within the set. Despite its complex appearance, the Mandelbrot set does not break into disjoint parts.

## Visualization

Visualizing the Mandelbrot set involves plotting the complex plane and iterating the function $ z_{n+1} = z_n^2 + c $ for each point $ c $. Points for which the sequence remains bounded are colored differently from those that diverge, creating a detailed and colorful fractal image.

## Significance in Mathematics

The Mandelbrot set is a central object in the study of complex dynamics. It provides insight into the behavior of iterative processes and serves as a rich source of examples and problems in fractal geometry and chaos theory. The study of the Mandelbrot set has led to discoveries in various fields of mathematics and science, illustrating the deep connections between simple iterative processes and complex structures.

## Python code for the Mandelbrot set is written in the following shell.




```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = c
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

def mandelbrot_set(xmin, xmax, ymin, ymax, width, height, max_iter):
    r1 = np.linspace(xmin, xmax, width)
    r2 = np.linspace(ymin, ymax, height)
    n3 = np.empty((width, height))

    for i in range(width):
        for j in range(height):
            n3[i, j] = mandelbrot(r1[i] + 1j*r2[j], max_iter)

    return (r1, r2, n3)

xmin, xmax, ymin, ymax = -2.0, 1.0, -1.5, 1.5
width, height = 1000, 1000
max_iter = 256

x, y, mandelbrot_image = mandelbrot_set(xmin, xmax, ymin, ymax, width, height, max_iter)

plt.imshow(mandelbrot_image.T, extent=[xmin, xmax, ymin, ymax], cmap='hot')
plt.colorbar()
plt.title('Mandelbrot Set')
plt.xlabel('Re')
plt.ylabel('Im')
plt.show()
```
